# Data read in, drop duplicates, and merging

```python
# Read in pre-COVID tweets, drop duplicates and save. 
raw_data_pre = pd.read_csv('./01_raw/pre_covid_learning.csv')
raw_data_pre = raw_data_pre.drop_duplicates()
raw_data_pre.to_csv('./01_raw/pre_covid_learning.csv')
```

```python
# Keep only the columns with information related to COVID or higher-education
cov_list = ['covid', 'covid-19', 'covid19', 'coronavirus', 'corona virus'] 
hied_list = ['college', 'university', 'professor', 'teacher', 'higher ed', 'higher education']
```

```python
# Convert all the text to lowercase for string matching
raw_data_pre['content'] = raw_data_pre['content'].str.lower()
```

```python
# Create a list of the indices containing Tweets with COVID or higher-education
int_keep_pre = []

for i in range(len(raw_data_pre)):
  val = str(raw_data_pre['content'][i])
  for j in hied_list:
    if j in val:
      if i in int_keep_pre:
        continue
      else:
        int_keep_pre.append(i)

int_keep_pre = pd.DataFrame(int_keep_pre)
int_keep_pre.columns = ['index']
int_keep_pre.to_csv('./02_intermediate/ints_pre.csv')
print("raw total:", len(raw_data_pre))

# Use the list of the indices to trim the raw df
df_pre = raw_data_pre.iloc[int_keep_pre['index']]
print("subset total: ", len(df_pre))
df_pre = df_pre.reset_index(drop=True)
df_pre.to_csv('./02_intermediate/clean_pre.csv')
```

```python
# Read in post-COVID tweets, drop duplicates and save
raw_data_post = pd.read_csv('./01_raw/post_covid_learning.csv')
print("Pre: ", len(raw_data_post))
raw_data_post = raw_data_post.drop_duplicates()
print("Post: ", len(raw_data_post))
raw_data_post.to_csv('./01_raw/post_covid_learning.csv')
```

```python
# Convert all the text to lowercase for string matching
raw_data_post['content'] = raw_data_post['content'].str.lower()
```

```python
# Create a list of the indices containing Tweets with COVID or higher-education
int_keep_post = []

for i in range(len(raw_data_post)):
  val = str(raw_data_post['content'][i])
  for j in hied_list:
    for k in cov_list:
      if j in val:
        if i in int_keep_post:
          continue
        else:
          int_keep_post.append(i)
      if k in val:
        if i in int_keep_post:
          continue
        else:
          int_keep_post.append(i)

int_keep_post = pd.DataFrame(int_keep_post)
int_keep_post.columns = ['index']
int_keep_post.to_csv('./02_intermediate/ints_post.csv')
print("raw total:", len(raw_data_post))

# Use the list of the indices to trim the raw df
df_post = raw_data_post.iloc[int_keep_post['index']]
print("subset total: ", len(df_post))
df_post = df_post.reset_index(drop=True)
df_post.to_csv('./02_intermediate/clean_post.csv')
```

```python
# Combine the pre and post dfs into a single dataframe
all_data = df_pre.append(df_post)
all_data = all_data.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
all_data = all_data.reset_index(drop=True)

# Add year col for later parsing
all_data['year'] = pd.DatetimeIndex(all_data['date']).year
all_data.to_csv('./02_intermediate/all_data.csv')
```

# Sentiment Analysis using Vader model

```python
# Add sentiment scores from Vader Model
san = SentimentIntensityAnalyzer()
pos = []
neu = []
neg = []
compound = []

for i in range(len(all_data)):
  sent_dict = san.polarity_scores(all_data['content'][i])
  pos.append(sent_dict.get('pos'))
  neu.append(sent_dict.get('neu'))
  neg.append(sent_dict.get('neg'))
  compound.append(sent_dict.get('compound'))

all_data['vader_pos'] = pos
all_data['vader_neu'] = neu
all_data['vader_neg'] = neg
all_data['vader_com'] = compound
```

```python
# Quick plots of Vader sentiment before and after
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
sns.violinplot(x=all_data['year'], y=all_data['vader_pos'], ax=axs[0, 0], palette="BrBG")
sns.violinplot(x=all_data['year'], y=all_data['vader_neu'], ax=axs[0, 1], palette="BrBG")
sns.violinplot(x=all_data['year'], y=all_data['vader_neg'], ax=axs[1, 0], palette="BrBG")
sns.violinplot(x=all_data['year'], y=all_data['vader_com'], ax=axs[1, 1], palette="BrBG")
plt.savefig(fname='./figures/violin_vader.png', bbox_inches='tight')
plt.show()
```

```python
# Descriptions of sentiment by year
vader_desc = pd.DataFrame(all_data['vader_pos'][all_data['year']==2020].describe())
vader_desc.columns = ['pos_2020']

vader_desc['pos_2021'] = all_data['vader_pos'][all_data['year']==2021].describe()

for i in ['vader_neu', 'vader_neg', 'vader_com']:
  for year in [2020, 2021]:
    name = i[-3:]+"_"+str(year)
    vader_desc[name] = all_data[i][all_data['year']==year].describe()

vader_desc.to_csv('./03_primary/vader_descriptions.csv')
vader_desc
```

```python
# Statistics on the differences
comparison = []
stat = []
p_val = []
coh_d = []

for i in ['vader_pos', 'vader_neu', 'vader_neg', 'vader_com']:
  comparison.append(i)
  ttest = stats.ttest_ind(all_data[i][all_data['year']==2020], all_data[i][all_data['year']==2021])
  stat.append(ttest[0])
  p_val.append(ttest[1])
  c0 = all_data[i][all_data['year']==2020]
  c1 = all_data[i][all_data['year']==2021]
  coh_d.append((mean(c0) - mean(c1)) / (sqrt((std(c0) ** 2 + std(c1) ** 2) / 2)))

vader_stats = pd.DataFrame([comparison, stat, p_val, coh_d])
vader_stats = vader_stats.transpose()
vader_stats.columns = ['comparison', 't_stat', 'p_val', 'cohens_d']
vader_stats.to_csv('./03_primary/vader_stats.csv')
vader_stats
```

```python
# Statistics on the differences
comparison = []
stat = []
p_val = []
coh_d = []

for i in ['text_blob_polarity', 'text_blob_subjectivity']:
  comparison.append(i)
  ttest = stats.ttest_ind(all_data[i][all_data['year']==2020], all_data[i][all_data['year']==2021])
  stat.append(ttest[0])
  p_val.append(ttest[1])
  c0 = all_data[i][all_data['year']==2020]
  c1 = all_data[i][all_data['year']==2021]
  coh_d.append((mean(c0) - mean(c1)) / (sqrt((std(c0) ** 2 + std(c1) ** 2) / 2)))

txblb_stats = pd.DataFrame([comparison, stat, p_val, coh_d])
txblb_stats = txblb_stats.transpose()
txblb_stats.columns = ['comparison', 't_stat', 'p_val', 'cohens_d']
txblb_stats.to_csv('./03_primary/txblb_stats.csv')
txblb_stats
```

```python
# Correlation matrix
def corrdot(*args, **kwargs):
    corr_r = args[0].corr(args[1], 'pearson')
    corr_text = f"{corr_r:2.2f}".replace("0.", ".")
    ax = plt.gca()
    ax.set_axis_off()
    marker_size = abs(corr_r) * 10000
    ax.scatter([.5], [.5], marker_size, [corr_r], alpha=0.6, cmap="coolwarm", \
               vmin=-1, vmax=1, transform=ax.transAxes)
    font_size = abs(corr_r) * 40 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction", ha='center', \
                va='center', fontsize=font_size)

def annotate_colname(x, **kws):
    ax = plt.gca()
    ax.annotate(x.name, xy=(0.05, 0.9), xycoords=ax.transAxes, fontweight='bold')

def plot_corr(df, title, save_name):
  sns.set(style='white', font_scale=1.6)
  g = sns.PairGrid(df, aspect=1.4, diag_sharey=False)
  g.fig.set_size_inches(20,20)
  g.map_lower(sns.regplot, scatter_kws={'s':5,'alpha':0.1})
  g.map_diag(sns.distplot, kde_kws={'color': 'black'})
  g.map_diag(annotate_colname)
  g.map_upper(corrdot)
  plt.title(title, fontsize=26) 
  plt.savefig(fname=f'./figures/{save_name}.png', bbox_inches='tight')
  plt.show()
```

```python
plot_corr(df=all_data[['vader_com', 'text_blob_polarity']], 
          title='All Data', 
          save_name='sentiment_corr_matrix')
plot_corr(df=all_data[['vader_com', 'text_blob_polarity']][all_data['year']==2020], 
          title='2020', 
          save_name='sent_corr_matrix_2020')
plot_corr(df=all_data[['vader_com', 'text_blob_polarity']][all_data['year']==2021], 
          title='2021', 
          save_name='sent_corr_matrix_2021')
```

# Topic Modeling

```python
# If picking up fresh
all_data = pd.read_csv('./03_primary/all_data.csv').drop(['Unnamed: 0'], axis=1)
```

## Removing stopwords and puncuation

```python
trim_data = all_data[['date', 'content', 'id', 'replyCount', 
                      'retweetCount', 'likeCount', 'quoteCount', 
                      'conversationId', 'lang', 'sourceLabel', 
                      'hashtags', 'time', 'year', 'vader_pos', 
                      'vader_neu', 'vader_neg', 'vader_com', 
                      'text_blob_polarity', 'text_blob_subjectivity']]
trim_data.head()
```

```python
# Removing stopwords
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
trim_data['content_no_stop'] = trim_data['content'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
trim_data['content_no_stop']
```

```python
# Removing punctuation
tokenizer = nltk.RegexpTokenizer(r"\w+")
clean = []
for i in range(len(trim_data)):
  sent = trim_data['content_no_stop'][i]
  clean.append(tokenizer.tokenize(sent))

trim_data['content_no_punct'] = clean
trim_data['content_no_punct']
```

## Unigrams

```python
# Frequency distributions - single word usage
# Common items that tend to stick around
drop_list = ["'", ',', '[', ']', "'t", "'co", "'https", "'s", "'m", "'de", 
             "'e", "'19", "'l", "'2", "'re", "'3", "'ve", "'4", "'5", "'eu", 
             "'don", "'na", "'da", "'30", "'pra", "'la", "'6", "'", "10", "'se", 
             "'w", "'ll", "'gt", "'12", "'q", "'8", "'7", "'d", "'a", "'u", "'00", 
             "'di", "'20", "'9", "'n", "'11", "'the", "'15", "'p", "'18", "'or", 
             "'ss", "'y", "'pq", "'ng", "'000", "'b", "'c", "'ga", "'16", "'50", 
             "'40", "'24", "'21", "'25", "'13", "'x", "'it", "'1"]

words_sent = []
for i in range(len(trim_data)):
  sent = str(trim_data['content_no_punct'][i])
  from nltk import word_tokenize
  sent = nltk.word_tokenize(sent)
  temp_sent = []
  for item in sent:
    if item in drop_list:
      continue
    else:
      item = item[1:]
      temp_sent.append(item)
  words_sent.append(temp_sent)

words = [item for sublist in words_sent for item in sublist] # Flatten words list

# Get frequency counts
fd = pd.DataFrame(nltk.FreqDist(words).most_common(1000))
fd.columns = ['word', 'values']
fd['word'] = fd['word'].str[1:]
top_100 = fd.iloc[:100, :]
```

```python
# Plot all
fig, ax = plt.subplots(figsize=(50,10))
sns.barplot(x=fd['word'], y=fd['values'], ax=ax)
plt.ylabel("Number of Occurrences", fontsize=20)
plt.savefig(fname='./figures/freq_dist_unigrams_1000.png', bbox_inches='tight')
plt.show()

# Plot top 100
fig, ax = plt.subplots(figsize=(10,20))
sns.barplot(y=top_100['word'], x=top_100['values'], ax=ax)
plt.ylabel("Number of Occurrences", fontsize=20)
plt.savefig(fname='./figures/freq_dist_unigrams_100.png', bbox_inches='tight')
plt.show()
```

```python
# Same plots but with search terms removed
remove_list = ['digital', 'learning', 'remote', 'delivery', 'teaching', 'online',
               'class', 'course', 'blended', 'synchronous', 'asynchronous', 
               'polysynchronous', 'hybrid', 'zoom', 'microsoft', 'teams', 
               'moodle', 'canvas', 'blackboard', 'webx' , 'google', 'classroom', 
               'covid', 'covid-19', 'covid19', 'coronavirus', 'corona virus', 
               'college', 'university', 'professor', 'teacher', 'higher', 'ed', 
               'education', 'microsoftteams', 'digitallearning', 'googleclassroom']

words_trim = [word for word in words if word not in remove_list]

# Get frequency counts
fd_t = pd.DataFrame(nltk.FreqDist(words_trim).most_common(100))
fd_t.columns = ['word', 'values']

# Plot top 100
fig, ax = plt.subplots(figsize=(10,20))
sns.barplot(y=fd_t['word'], x=fd_t['values'], ax=ax)
plt.ylabel("Number of Occurrences", fontsize=20)
plt.savefig(fname='./figures/freq_dist_unigrams_100_search_removed.png', bbox_inches='tight')
plt.show()
```

## Bigrams

```python
from nltk.util import ngrams
# Frequency distributions - bigrams
bgs_ind = []

# Convert each sentence into bigrams
for sentence in words_sent:
  bgs_ind.append(list(ngrams(sentence, 2)))

# Get a single list of all the bigrams
bgs = []
for i in range(len(bgs_ind)):
  sent_tuples = bgs_ind[i]
  for tup in sent_tuples:
    bgs.append(tup)

# Get frequency counts
fd_bgs = pd.DataFrame(nltk.FreqDist(bgs).most_common(1000))
fd_bgs.columns = ['bigram', 'values']
top_100_bgs = fd_bgs.iloc[:100, :]
```

```python
# Plot top 100
fig, ax = plt.subplots(figsize=(10,20))
sns.barplot(y=top_100_bgs['bigram'], x=top_100_bgs['values'], ax=ax)
plt.ylabel("Number of Occurrences", fontsize=20)
plt.xticks(rotation=90)
plt.savefig(fname='./figures/freq_dist_bigrams_100.png', bbox_inches='tight')
plt.show()
```

```python
# Plot top 100 with search terms removed
bgs_t = []

for tup in bgs:
  append = True
  for word in tup:
    if word in remove_list:
      append=False
    else:
      continue
  if append==True:
    bgs_t.append(tup)

fd_bgs_t = pd.DataFrame(nltk.FreqDist(bgs_t).most_common(100))
fd_bgs_t.columns = ['bigram', 'values']

# Plot it
fig, ax = plt.subplots(figsize=(10,20))
sns.barplot(y=fd_bgs_t['bigram'], x=fd_bgs_t['values'], ax=ax)
plt.ylabel("Number of Occurrences", fontsize=20)
plt.xticks(rotation=90)
plt.savefig(fname='./figures/freq_dist_bigrams_100_search_removed.png', bbox_inches='tight')
plt.show()
```

## Trigrams

```python
from nltk.util import ngrams
# Frequency distributions - trigrams
tgs_ind = []

# Convert each sentence into trigrams
for sentence in words_sent:
  tgs_ind.append((ngrams(sentence, 3)))

# Get a single list of all the trigrams
tgs = []
for i in range(len(tgs_ind)):
  sent_tuples = list(tgs_ind[i])
  for tup in sent_tuples:
    tgs.append(tup)

# Get frequency counts
fd_tgs = pd.DataFrame(nltk.FreqDist(tgs).most_common(1000))
fd_tgs.columns = ['trigram', 'values']
top_100_tgs = fd_tgs.iloc[:100, :]
```

```python
# Plot top 100 with search terms removed
tgs_t = []

for tup in tgs:
  append = True
  for word in tup:
    if word in remove_list:
      append=False
    else:
      continue
  if append==True:
    tgs_t.append(tup)

fd_tgs_t = pd.DataFrame(nltk.FreqDist(tgs_t).most_common(100))
fd_tgs_t.columns = ['trigram', 'values']

# Plot it
fig, ax = plt.subplots(figsize=(10,20))
sns.barplot(y=fd_tgs_t['trigram'], x=fd_tgs_t['values'], ax=ax)
plt.ylabel("Number of Occurrences", fontsize=20)
plt.xticks(rotation=90)
plt.savefig(fname='./figures/freq_dist_trigrams_100_search_removed.png', bbox_inches='tight')
plt.show()
```
